### Exercise 1  

Write a function called `distance_between_points` that takes two `Points` as arguments and returns the distance between them.

````python
import math
def distance_between_points(point1, point2):
    """
    computes the distance between two points, where the points are 
    represented as lists
    
    >>> distance_between_points([0, 1], [0, 2])
    1.0
    
    >>> distance_between_points([1, 1], [0, 2])
    1.4142135623730951
    
    >>> distance_between_points([1, 1], [1, 1])
    0.0
    """
    
    distance = math.sqrt((point1[0]-point2[0])**2 + (point1[1] - point2[1])**2)
    return distance
````

### Exercise 2  

Write a function named `move_rectangle` that takes a `Rectangle` and two numbers named `dx` and `dy`. It should change the location of the rectangle by adding `dx` to the `x` coordinate of `corner` and adding `dy` to the `y` coordinate of `corner`.

````python
class Rectangle(object):
    """Represents a rectangle. 

    attributes: width, height, corner.
    """

class Point(object):
    """
    Represents a point - the corner of the rectangle
    """

box = Rectangle()
box.width = 100.0
box.height = 200.0
box.corner = Point()
box.corner.x = 0.0
box.corner.y = 0.0

def move_rectangle(rect, dx, dy):
    """
    changes the origin point of the start of the recangle by adding
    dx and dy to the corrdinates of the original point
    
    """
    box.corner.x = box.corner.x + dx
    box.corner.y = box.corner.y + dy
    return rect

move_rectangle(box, 1, 1)
print box.corner.y
````

### Exercise 3  

Write a version of `move_rectangle` that creates and returns a new `Rectangle` instead of modifying the old one.

````python
class Rectangle(object):
    """Represents a rectangle. 

    attributes: width, height, corner.
    """

class Point(object):
    """
    Represents a point - the corner of the rectangle
    """

box = Rectangle()
box.width = 100.0
box.height = 200.0
box.corner = Point()
box.corner.x = 0.0
box.corner.y = 0.0

def move_rectangle(rect, dx, dy):
    """
    Returns a new rectange with a new origin point, defined by the 
    old rectangle's origin point plus dx and dy
    """
    r = Rectangle()
    r.width = rect.width
    r.height = rect.height
    r.corner = Point()
    r.corner.x = box.corner.x + dx
    r.corner.y = box.corner.y + dy
    return r
    
````
